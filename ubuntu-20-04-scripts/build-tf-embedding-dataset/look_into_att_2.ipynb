```python
import pickle
import os
```

```python
def get_pickle_file_content(full_path_pickle_file):
    pickle_file = open(full_path_pickle_file,'rb')
    pickle_list = pickle.load(pickle_file, encoding='latin1')
    pickle_file.close()
    
    return pickle_list
```

```python
#### main
full_path_pickle_file = "/tmp/savetest/att-tokenized-abigail-tools.pickle"

pickle_list = get_pickle_file_content(full_path_pickle_file)

pickle_list_item = next(iter(pickle_list))
print(f'item-0: {pickle_list_item[0]}')
print(f'item-1: {pickle_list_item[1]}')

c = 0
for dis,ret in pickle_list:
    print(f'dis:{dis}')
    c +=1
    
    if c > 0:
        break
```
