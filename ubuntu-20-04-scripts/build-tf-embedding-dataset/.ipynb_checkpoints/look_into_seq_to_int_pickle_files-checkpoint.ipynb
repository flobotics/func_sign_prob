```python
import pickle
import os
import tarfile
```

```python
def untar_one_pickle_file(full_path_tar_file, work_dir):
    tar = tarfile.open(full_path_tar_file, "r:bz2")  
    tar.extractall(work_dir)
    tar.close()
```

```python
def get_pickle_file_content(full_path_pickle_file):
    pickle_file = open(full_path_pickle_file,'rb')
    pickle_list = pickle.load(pickle_file, encoding='latin1')
    pickle_file.close()
    
    return pickle_list
```

```python
### main
filename = "/tmp/embbuild/att-int-seq-abigail-tools.pickle.tar.bz2"
#filename_pickle = "/tmp/embtest/att-int-seq-abigail-tools.pickle"

filename_pickle = '/tmp/embstoredir/full_dataset_att_int_seq.pickle'


#untar_one_pickle_file(filename, '/tmp/embtest')

content = get_pickle_file_content(filename_pickle)
#item = next(iter(content))
#print(f'content: {item}')

c = 0

for i in content:
    for dis,ret in i:
        print(f'dis:{dis}  ret:{ret}')

    c += 1
    if c > 3:
        break
```
